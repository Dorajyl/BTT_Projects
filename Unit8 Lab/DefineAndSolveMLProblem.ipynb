#### 2. Training and evaluating model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30,  random_state=1234)
```

##### Linear Regression Model

```python
# building linear model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# get prediction
y_lr_pred = lr_model.predict(X_test)

# root mean sqaure error and r2 for evaluation
lr_rmse = mean_squared_error(y_test, y_lr_pred, squared=False)
lr_r2 = r2_score(y_test, y_lr_pred)
print('[LR] Root Mean Squared Error: {0}'.format(lr_rmse))
print('[LR] R2: {0}'.format(lr_r2))
```

##### Decision Tree Model

```python
# set up parameter grid
param_grid = {'max_depth': [4,8,12], 'min_samples_leaf': [25, 50, 75]}
param_grid
```

```python
print('Running Grid Search...')

# create a DecisionTreeRegressor model object without supplying arguments. 
dt_regressor = DecisionTreeRegressor()

# run a Grid Search with 3-fold cross-validation 
dt_grid = GridSearchCV(dt_regressor, param_grid, cv = 3, scoring='neg_root_mean_squared_error')

# fit the model (use the 'grid' variable) on the training data 
dt_grid_search = dt_grid.fit(X_train, y_train)

print('Done')
```

```python
# get best rmse
rmse_DT = -1 * dt_grid_search.best_score_
print("[DT] RMSE for the best model is : {:.2f}".format(rmse_DT) )
```

```python
# get best parameters
dt_best_params = dt_grid_search.best_estimator_
dt_best_params
```

```python
# best decision tree model
dt_model = DecisionTreeRegressor(max_depth=8, min_samples_leaf=50)
dt_model.fit(X_train, y_train)
# use the fitted model to make predictions on the test data
y_dt_pred = dt_model.predict(X_test)

# compute the RMSE 
dt_rmse = mean_squared_error(y_test, y_dt_pred, squared=False)

# compute the R2 score
dt_r2 = r2_score(y_test, y_dt_pred)

print('[DT] Root Mean Squared Error: {0}'.format(dt_rmse))
print('[DT] R2: {0}'.format(dt_r2))
```

#### 3. Model Selection

```python
RMSE_Results = [lr_rmse, dt_rmse]
R2_Results = [lr_r2, dt_r2]

rg= np.arange(2)
width = 0.35

# 1. Create bar plot with RMSE results
# YOUR CODE HERE
plt.bar(rg-0.17, RMSE_Results, color ='blue', 
        width = width, label = "RMSE")

# 2. Create bar plot with R2 results
# YOUR CODE HERE
plt.bar(rg+0.17, R2_Results, color ='green', 
        width = width, label = "R2")


labels = ['LR', 'DT']
plt.xticks(rg + width/2, labels)

plt.xlabel("Models")
plt.ylabel("RMSE/R2")


plt.ylim([0,0.7])
plt.title('Model Performance')
plt.legend(loc='upper left', ncol=2)
plt.show()
```

From the above, we can see that the model performance are almost identical from the linear regression and decision tree models. As a result, I decided to incorporate other ensemble models and see their performance incomparison.

##### Stacking Emsemble

```python
estimators = [("DT", DecisionTreeRegressor(max_depth=8, min_samples_leaf=50)), # parameter from grid search
              ("LR", LinearRegression())
             ]
```

```python
print('Implement Stacking...')

stacking_model = StackingRegressor(estimators = estimators, passthrough=False)
stacking_model.fit(X_train, y_train)

print('End')
```

```python
# get model prediction
stacking_pred = stacking_model.predict(X_test)

# calculate scores
stack_rmse = mean_squared_error(y_test, stacking_pred, squared=False)
stack_r2 = r2_score(y_test, stacking_pred) 
print('Root Mean Squared Error: {0}'.format(stack_rmse))
print('R2: {0}'.format(stack_r2)) 
```

Comparing the result of stacking to linear regression and decision tree

```python
RMSE_Results = [lr_rmse, dt_rmse, stack_rmse]
R2_Results = [lr_r2, dt_r2, stack_r2]

rg= np.arange(3)
width = 0.35

# 1. Create bar plot with RMSE results
# YOUR CODE HERE
plt.bar(rg-0.17, RMSE_Results, color ='blue', 
        width = width, label = "RMSE")

# 2. Create bar plot with R2 results
# YOUR CODE HERE
plt.bar(rg+0.17, R2_Results, color ='green', 
        width = width, label = "R2")

labels = ['LR', 'DT', 'Stacking']
plt.xticks(rg + width/2, labels)

plt.xlabel("Models")
plt.ylabel("RMSE/R2")

plt.ylim([0,0.7])
plt.title('Model Performance')
plt.legend(loc='upper left', ncol=2)
plt.show()
```

After modifying the y axis range to 0 to 0.7, we can see that decision tree and stacking are performing slightly better than the linear regression model. However, the different is very slight.
